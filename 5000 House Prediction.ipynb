# For Fast Evaluation

```python
def training_evaluation (actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2_score: {r2}")
    plt.scatter(Y_train, y_pred_train, alpha=0.5, color='blue', label = "Actual Price")
    plt.plot([Y_train.min(), Y_train.max()], [Y_train.min(), Y_train.max()] , 
         color='k', linestyle='dashed', linewidth=3, label='Predicted Price')
    plt.title('Scatter PLot on Training Data')
    plt.xlabel("Actual Price")
    plt.ylabel("Predicted Price")
    plt.legend()

```

```python
training_evaluation(Y_train,y_pred_train)
```

# For Test Data

```python
def training_evaluation_test (actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)
    
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")
    print(f"R2_score: {r2}")
    plt.scatter(actual, predicted, alpha=0.5, color='blue', label = "Actual Price")
    plt.plot([actual.min(), actual.max()], [actual.min(), actual.max()] , 
         color='k', linestyle='dashed', linewidth=3, label='Predicted Price')
    plt.title('Scatter PLot on Test Data')
    plt.xlabel("Actual Price")
    plt.ylabel("Predicted Price")
    plt.legend()

```

```python
training_evaluation_test(Y_test, y_pred_test)
```

```python
residuals = Y_test - y_pred_test
sns.histplot(residuals, kde=True)
plt.title("Residual Distribution")
plt.show()
```
